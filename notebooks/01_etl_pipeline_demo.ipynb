```python
# Basic application bootstrap cell

APP_NAME = "DataBricks-Sample1"
ENV = "dev"

settings = {
    "input_path": "/tmp/input",
    "output_path": "/tmp/output",
    "run_mode": "batch",
}

def run_health_check(config: dict) -> bool:
    required_keys = {"input_path", "output_path", "run_mode"}
    return required_keys.issubset(config.keys())

is_ready = run_health_check(settings)

print(f"Starting {APP_NAME} in {ENV} mode")
print(f"Configuration valid: {is_ready}")
```
